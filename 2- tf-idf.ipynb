```python
#import libraries
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer

#örnek belge oluştur
documents = ["köpek çok tatlı bir hayvandır"
             ,"köpek ve kuşlar çok tatlı bir hayvandır."
             ,"inekler süt üretirler"]

#vectorizer tanımla
tfidf_vectorizer = TfidfVectorizer()
```

```python
#metinleri sayısal hale çevirme
X = tfidf_vectorizer.fit_transform(documents)

#kelime kümesini incele
feature_names = tfidf_vectorizer.get_feature_names_out()

#vektör temsilini incele
vektor_temsili= X.toarray()
vektor_temsili
```

```python
df_tfidf = pd.DataFrame(vektor_temsili,columns=feature_names)
df_tfidf
```

```python
#ortalama tf-idf değerlerine bakalım
tf_idf= df_tfidf.mean(axis=0)
tf_idf
```

```python
#import libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

#veri setini yükleyelim
df = pd.read_csv("spam.csv", encoding='cp1252')
df
```

```python
#veri temizleme
```

```python
#tfidf
vectorizer= TfidfVectorizer()
X =vectorizer.fit_transform(df.v2)
```

```python
#kelime kümesini inceleyelyim
feature_names = vectorizer.get_feature_names_out()
tfidf_score= X.mean(axis=0).A1 # her kelimeninm ortalama tf-idf değerleri
```

```python
#tfidf-scorelarını içeren bir df oluştur
df_tfidf = pd.DataFrame({"word":feature_names,"tfidf_score":tfidf_score})
df_tfidf
```

```python
#skorları sırala ve sonucu incele
df_tfidf_sorted = df_tfidf.sort_values(by = "tfidf_score",ascending=False)
df_tfidf_sorted
```
