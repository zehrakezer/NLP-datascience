```python
#metin temizleme ve processing  : tokenization, padding, label encoding
#tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text) #metinler üzeirndeki frekansı öğren fit et
total_words = len(tokenizer.word_index)+1 #toplam kelime sayısı

#n-gram dizileri oluştur ve padding uygula
input_sequences= []
for texts in text:
    #metinleri kelime indexlerşne çevri
    token_list =tokenizer.texts_to_sequences([texts])[0]

    #her metin için n-gram dizisi oluşturalım
    for i in range(1,len(token_list)):
        n_gram_seq =token_list[:i+1]
        input_sequences.append(n_gram_seq)
```

```python

#en uzun diziyi bulalım, tüm diileri aynı uzualuğa getirelim
max_sequence_length= max(len(x) for x in input_sequences)

#dizilere padding işlemi uygula, hepsinin aynı uzunlukta olmasını sağla
input_sequences = pad_sequences(input_sequences,maxlen=max_sequence_length,padding="pre")

#X(girdi) ve y(hedef değişken)
X = input_sequences[:,:-1]
y = input_sequences[:,-1]

y=tf.keras.utils.to_categorical(y,num_classes=total_words) #one-hot encoding
```

```python
#LSTM modeli oluştur compile, train, ve evaluate
model =Sequential()

#embedding
model.add(Embedding(total_words,50,input_length= X.shape[1]))

#lstm 
model.add(LSTM(100,return_sequences=False))

#output 
model.add(Dense(total_words,activation="softmax"))

#model compile
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
#model training 
model.fit(X,y, epochs=100,verbose=1)
```

```python
#model prediction
def generate_text(seed_text,next_words):

    for _ in range(next_words):

        #girdi merinlerini sayısal verilere dönüştür
        token_list = tokenizer.texts_to_sequences([seed_text])[0]

        #padding
        token_list=pad_sequences([token_list], maxlen=max_sequence_length-1,padding="pre")

        #prediction
        predicted_probabilities = model.predict(token_list, verbose=0)

        #en yüksek olasılığa sahip kelimenin indexini bul
        predicted_word_index = np.argmax(predicted_probabilities,axis =-1)

        #tokenizer ile kelime indexinden asıl kelimeyi bul
        predşicted_word=tokenizer.index_word[predicted_word_index[0]]

        #tahmin edilen kelimeyi sed_texte ekleyelim
        seed_text= seed_text +" "+ predşicted_word

    return seed_text
```
