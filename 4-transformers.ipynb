```python
#import libraries
from transformers import AutoTokenizer, AutoModel
import torch

#model ve tokenizer y√ºkle
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
#input text metni tanƒ±m≈üa
text = "Transformers can be used for natural language processing"

#metni tokenlere √ßevirm
inputs = tokenizer(text, return_tensors="pt") #√ßƒ±ktƒ± pytorch tensoru olarka return edilir

#modeli kullanarak metin temsili olu≈ütur
with torch.no_grad():#gradyanlarƒ±n hesaplanmasƒ± durdurulur, b√∂ylece belleƒüi daha verimli kullanƒ±rƒ±z
    outputs = model(**inputs)

#modelin √ßƒ±kƒ±≈üƒ±ndan son gizli durumu alalƒ±m
last_hidden_state = outputs.last_hidden_state #t√ºm token √ßƒ±ktƒ±larƒ±nƒ± alamk i√ßin

#ilk token embeddingi alalƒ±m ve print edelim
first_token_embedding = last_hidden_state[0,0,:].numpy()
first_token_embedding
```
